```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
topics = pd.read_csv("topic_with_desc.csv")
# Topics, Description, Keywords

bucket = pd.read_csv("net_bucket.csv")
# Bucket Number, Bucket
```

```python
print(type(topics["Topics"]))

# for topic in topics["Topics"]:
#     print(topic)


for idx in topics.index:
    print(topics["Topics"][idx])
    print(topics["Description"][idx]) 
    print(topics["Keywords"][idx])
    print()
```

```python
# Create a TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Create a list of all topic descriptions and keywords from BERTopic and book topics
all_descriptions = []
for idx in topics.index:
    # Assign a higher weight to keywords by repeating them multiple times
    description = topics["Description"][idx]
    keywords = " ".join(topics["Keywords"][idx].split(',')) * 3  # Repeat keywords 3 times for higher weight
    all_descriptions.append(f"{description} {keywords}")
    

# Fit and transform the TF-IDF vectorizer on all descriptions
tfidf_matrix = tfidf_vectorizer.fit_transform(all_descriptions)

# Initialize an empty list to store topic mappings
topic_mappings = []

# Iterate through each list of ngrams in bucket_ngrams
for ngram_list in bucket["Bucket"].tolist():
    ngram_string = " ".join(ngram_list)

    # Calculate cosine similarity between the list of ngrams and book topics
    cosine_similarities = cosine_similarity(tfidf_vectorizer.transform([ngram_string]), tfidf_matrix)

    # Find the book topic with the highest similarity
    similarity_scores = cosine_similarities[0]
    book_topic_index = cosine_similarities.argmax()
    book_topic = topics["Topics"][book_topic_index]
    
    topic_mappings.append({
        "bucket": ngram_list,
        "topic": book_topic,
        "similarity": similarity_scores
    })
    

# Print the topic mappings for each list of ngrams
for i, ngram_list_mapping in enumerate(topic_mappings):
    print(f"{ngram_list_mapping['bucket']} maps to book topic \n {ngram_list_mapping['topic']} with similarity score \n{ngram_list_mapping['similarity']}\n")
```
