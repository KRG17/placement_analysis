```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
topics = pd.read_csv("topic_with_desc.csv")
# Topics, Description, Keywords

bucket = pd.read_csv("net_bucket.csv")
# Bucket Number, Bucket
```

```python
print(type(topics["Topics"]))

# for topic in topics["Topics"]:
#     print(topic)


for idx in topics.index:
    print(topics["Topics"][idx])
    print(topics["Description"][idx]) 
    print(topics["Keywords"][idx])
    print()
```

```python
tfidf_vectorizer = TfidfVectorizer()

all_description = []
for idx in topics.index:
    description = topics["Description"][idx]
    keywords = " ".join(topics["Keywords"][idx].split(','))
    keywords = keywords * 3
    all_description.append(f"{description} {keywords}")

tfidf_matrix = tfidf_vectorizer.fit_transform(all_description)

# print(tfidf_matrix)

mapping = []

for idx in bucket.index:
    bucket_topic = bucket["Bucket"][idx]
    bucket_topic = " ".join(bucket_topic.split(','))
    cosine_similarities = cosine_similarity(tfidf_vectorizer.transform([bucket_topic]), tfidf_matrix)

#     print(cosine_similarities)

    book_topic_index = cosine_similarities.argmax()
    book_topic = topics["Topics"][book_topic_index]
    
    mapping.append({
        "bucket_topic_number": idx,
        "bucket_topic": bucket["Bucket"][idx],     #bucket_topic
        "book_topic_number": cosine_similarities.argmax(),
        "book_topic": book_topic
    })

    print(bucket_topic)
    print(book_topic)
    print()

df = pd.DataFrame(mapping)

df.to_csv("mapping.csv", index=False)
```
